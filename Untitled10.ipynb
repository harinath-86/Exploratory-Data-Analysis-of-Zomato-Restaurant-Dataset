```python
# Top Rated restraunts city wise
#Low Rated restraunts city wise
```

```python
top_rated=data[data['Aggregate rating']>4.5]
```

```python
low_rated=data[data['Aggregate rating']<3.4]
```

```python
figure=px.bar(top_rated,x='City',y='Aggregate rating')
figure.show()
```

```python
figure=px.scatter(low_rated,x='City',y='Aggregate rating',color='Rating color')
figure.show()
```

```python
#City wise analysis how many City has Online delivery System
data.groupby('City').agg('Has Online delivery').value_counts().sort_values(ascending=False).unstack().fillna(0).astype(int).sort_values(['No','Yes'],ascending=False)
```

```python
#City wise analysis How many city has table booking feature
a=data.groupby('City').agg('Has Table booking').value_counts().sort_values(ascending=False)

a.unstack().fillna(0).astype(int).sort_values(['No','Yes'],ascending=False).reset_index()
```

```python
#top rated restraunt city wise on map
fig = px.scatter_mapbox(top_rated, lat="Latitude", lon="Longitude", hover_name="City", hover_data=["Aggregate rating", "Restaurant Name"],
                        color_discrete_sequence=["fuchsia"], zoom=4, height=300)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.update_layout(title='Highly rated Resturants Location',
                  autosize=True,
                  hovermode='closest',
                  showlegend=False)
fig.update_layout(
    autosize=False,
    width=800,
    height=500,)

fig.show()
```

```python
#map of new delhi where north indian cuisines is eaten the most (zoom in to see the places in delhi)
delhi=data[(data['Cuisines']=='North Indian') & (data['City']=='New Delhi')]
fig = px.scatter_mapbox(delhi, lat="Latitude", lon="Longitude", hover_name="City", hover_data=["Cuisines", "Restaurant Name"],
                        color_discrete_sequence=["fuchsia"], zoom=2, height=300)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.update_layout(title='North Indian Cuisines in delhi',
                  autosize=True,
                  hovermode='closest',
                  showlegend=False)
fig.update_layout(
    autosize=False,
    width=800,
    height=500,)

fig.show()
```
